```python
import numpy as np
import pandas as pd

data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.tail(2)

# Set the maximum number of rows displayed when printing DataFrames
pd.set_option("display.max_rows", 30)

# Print the DataFrame with the set maximum number of rows
df
```
